# Apply Harris Corner Detection

1-Conver the gray image dtype to float32 instead of being in uint8

```python
gray= np.float32(gray_flat_chess)
```

2-Apply cornerHarris to the modified image

```python
dst = cv2.cornerHarris(src=gray, blockSize=2 , ksize=3 , k= 0.04)
```

3- Dilate the result of cornerHarris to help for visualizing the image

```python
dst = cv2.dilate(dst,None)
```

4-Mark the detected corners in red

```python
flat_chess[dst>(0.01*dst.max())]=[255,0,0] 
```

5-Show the image

```python
plt.imshow(flat_chess)
```

```python
gray= np.float32(gray_real_chess)
dst = cv2.cornerHarris(src=gray, blockSize=2 , ksize=3 , k= 0.04)
dst = cv2.dilate(dst,None)
real_chess[dst>(0.01*dst.max())]=[255,0,0]
plt.imshow(real_chess)
```

# Shi-Tomasi Corner Detection

```python
flat_chess= cv2.imread("D:\Empty_flat_chess.jpg")
flat_chess= cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)

real_chess = cv2.imread("D:\chess_board.jpg")
real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2RGB)

gray_flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)

gray_real_chess= cv2.cvtColor(real_chess,cv2.COLOR_BGR2GRAY)
```

```python
corners = cv2.goodFeaturesToTrack(gray_flat_chess, 60 , 0.01 , 10 ) #--> max No. of corners to return = 80 , -1 for all corners
                                                                    #--> euclidean distance = 10
```

```python
corners= np.int0(corners) # convert corners from float dtype to integer dtype (int0 if we are not sire about dtype)
```

```python
corners
```

The numpy.ravel() functions returns contiguous flattened array(1D array with all the input-array elements and with the same type as it). A copy is made only if needed. 

```python
for i in corners:
    x,y= i.ravel()
    #print( x)
    #print (y)
    #print('*************************************************************************')
    cv2.circle(flat_chess, (x,y) , 5 , (255,0,0) , -1 )
```
