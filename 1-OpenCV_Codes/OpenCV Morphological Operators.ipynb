```python
import pandas as pd 
import numpy as np
import cv2 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img():
    blank_img= np.zeros((600,600))
    font= cv2.FONT_HERSHEY_COMPLEX
    cv2.putText(blank_img,text='ABCDE',org=(50,300),fontFace= font,fontScale=5, color=(255,255,255) ,thickness=20)
    return blank_img
```

```python
def display_img(img):
    fig=plt.figure(figsize=(12,10))
    ax=fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
img= load_img()
display_img(img)
```

```python
kernel= np.ones((5,5), dtype=np.uint8)
```

```python
kernel
```

```python
result = cv2.erode(img, kernel, iterations=4)
```

```python
display_img(result)
```

```python
white_noise= white_noise * 255
```

```python
noise_img = white_noise + img
```

```python
display_img(noise_img)
```

```python
############################ Use Openning (erosion -> Dilation) to get rid of this noise ###########################
```

```python
openning = cv2.morphologyEx(noise_img, cv2.MORPH_OPEN , kernel) # pass--> src , operation , kernel 
```

```python
display_img(openning)
```
