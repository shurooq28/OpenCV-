```python
import pandas as pd
import numpy as np 
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("D:/Sammy_dog_face.jpg")
```

```python
plt.imshow(img)
```

```python
 edges = cv2.Canny(image=img , threshold1=127, threshold2=127) # Hysteresis Thresholding
```

```python
plt.imshow(edges)
```

```python
# Edges is not clear and to adjust this we can 1-Blur the img  2-playying around threshold1 and threshold2
```

```python
edges = cv2.Canny(image=img , threshold1=0, threshold2=255)
plt.imshow(edges)
```

```python
# Formula to adjust threshold1 and threshold2
```

```python
#1- Calculate the median pixel value 
```

```python
med_val= np.median(img)
```

```python
med_val
```

```python
# Set lower threshold to either 0 or 70% of median value whichever is greater
lower = int(max(0, 0.7*med_val))                               #threshold1
# Set lower threshold to either 130% of median value or max 255 whichever is smaller 
upper = int(min(0, 1.3*med_val))                              #threshold2
```

```python
edges = cv2.Canny(image=img , threshold1=lower+30, threshold2=upper+150)
plt.imshow(edges)
```

```python
blur_img=cv2.blur(img,ksize=(5,5))
```

```python
plt.imshow(blur_img)
```

```python
edges = cv2.Canny(image=blur_img , threshold1=lower+5, threshold2=upper-10)
plt.imshow(edges)
```
