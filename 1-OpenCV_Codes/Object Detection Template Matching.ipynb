```python
# All the 6 methods for comparison on a list from the ducumentation of matchTemplate
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR',
            'cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
```

```python
for m in methods:
    
#Creat A copy of the image
    
    full_copy= full.copy()
    
    method = eval(m)
    
#TEMPLATE MATCHING
    
    result = cv2.matchTemplate(full_copy, face , method)
    
# Determine result's top_left to used it for drawing rectangle
    
    min_val, max_val , min_loc, max_loc=cv2.minMaxLoc(result)
    
    if method in ['cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']:
        top_left= min_loc  #(x,y)
    else:
        top_left=max_loc
        
# Using Result top_left to determine bottom_right of retangle from matching template
    
    hight,width,channels= face.shape
    
    bottom_right= (top_left[0]+width , top_left[1]+hight)
    
    cv2.rectangle(full_copy, top_left , bottom_right, (255,0,0), 10)
    
# PLOT AND SHOW IMGES
    
    plt.subplot(121)    # one row and two columns
    plt.imshow(result)
    #plt.title('HEATMAP OF TEMPLATE MATCHING')
    
    plt.subplot(122)   
    plt.imshow(full_copy)
    #plt.title('DETECTION OF TEMPLATE')
    #plt.suptitle(m)
    
    
    plt.show()
    
    print('/n')
print("/n")
```
