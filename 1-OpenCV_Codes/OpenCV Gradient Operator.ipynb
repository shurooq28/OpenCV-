```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
```

```python
def display_img(img):
    fig=plt.figure(figsize=(12,10))
    ax=fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
img = cv2.imread("D:\Sudoku.jpg",0)
```

```python
display_img(img)
```

```python
sobelx= cv2.Sobel(img, cv2.CV_64F , 1, 0 , ksize=5)  #  cv2.CV_64F --> img depth (precision of each pixel) 
```

```python
# x-Gradient --> emerge the vertical lines of image 
```

```python
display_img(sobelx)
```

```python
sobely= cv2.Sobel(img, cv2.CV_64F , 0, 1 , ksize=5)
```

```python
display_img(sobely)
```

```python
                             ########## Laplace Operator ####################
```

```python
laplacian= cv2.Laplacian(img , cv2.CV_64F)
```

```python
display_img(laplacian)
```

```python
###########################################################################################################
```

```python
blended = cv2.addWeighted(src1=sobelx , alpha=0.5 , src2=sobely, beta=0.5 , gamma=0)
```

```python
display_img(blended)
```

```python
ret , th1 = cv2.threshold(img , 100 ,255 , cv2.THRESH_BINARY )
```

```python
display_img(th1)
```

```python
################################################################################################
      ##########    Morphological gradient for edge detection ###########
################################################################################################
```

```python
kernel= np.ones((4,4),dtype= np.uint8)
```

```python
gradient = cv2.morphologyEx(blended , cv2.MORPH_GRADIENT , kernel )
```

```python
display_img(gradient)
```
