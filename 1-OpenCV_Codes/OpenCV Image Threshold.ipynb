```python
img2= cv2.imread("D:\Crossword-Puzzle.jpg",0)
plt.imshow(img2, cmap='gray')
```

```python
def show_img(img):
    fig=plt.figure(figsize=(15,15))
    ax=fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
show_img(img2)
```

```python
ret, th1 = cv2.threshold(img2, 50, 255, cv2.THRESH_BINARY)
show_img(th1)
```

```python
ret, th1 = cv2.threshold(img2, 180, 255, cv2.THRESH_BINARY)
show_img(th1)
```

```python
# There are two ways to prevent fadded black due to threshold 
# 1_st: change the threshold level or number as above
# 2_nd: use cv2.adaptiveThreshold and choose adaptiveGuassian or adaptiveMean
```

```python
th2 = cv2.adaptiveThreshold(img2,255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY , 5,5)
show_img(th2)
```

```python
blended = cv2.addWeighted(src1= th1 , alpha= 0.6 , src2= th2, beta= 0.8 , gamma= 0)  
show_img(blended)
```
