```python
# Resize Images : We have two ways to resize an image:* By Dimensions 
#                                                     * By Ratio
```

```python
# Resize By Dimensions

resized_img = cv2.resize(fix_img,(1800,500)) # Width= 1800  ,Height= 500  order is swapped 
plt.imshow(resized_img)
```

```python
# Resize By Ratios

w_ratio = 0.8
h_ratio = 0.2  # 20% of original height

new_img = cv2.resize(fix_img, (0,0) , fix_img , w_ratio , h_ratio)

```

```python
plt.imshow(new_img)
```

```python
new_img.shape
```

```python
# Image Flipping 
```

```python
flip_img = cv2.flip(fix_img,0) # to flip image horizontally
```

```python
plt.imshow(flip_img)
```

```python
flip_img = cv2.flip(fix_img,1)
```

```python
plt.imshow(flip_img) # to flip image vertically
```

```python
flip_img = cv2.flip(fix_img,-1)  # to flip image horizontally and vertically
plt.imshow(flip_img)                
```
